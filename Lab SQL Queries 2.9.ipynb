```python
#How many distinct (different) actors' last names are there?
```

```python
distinct_names = pd.read_sql_query("SELECT COUNT(DISTINCT(last_name)) AS Distinct_Names FROM sakila.actor;", engine)

distinct_names = pd.DataFrame(distinct_names)
```

```python
#Since how many days has the company been operating (check DATEDIFF() function)? 
```

```python
days_operating = pd.read_sql_query("SELECT MAX(payment_date) AS 'FIRST DAY', MIN(payment_date) AS 'LAST DAY', DATEDIFF('2006-02-14', '2005-05-24') AS 'DAYS OPERATING' FROM sakila.payment;", engine)
days_operating = pd.DataFrame(days_operating)
```

```python
days_operating.head()
```

```python
#Show rental info with additional columns month and weekday. Get 20 results. 
```

```python
rental_info = pd.read_sql_query("SELECT *, MONTH(rental_date) AS 'month', DAY(rental_date) AS 'weekday' FROM sakila.rental LIMIT 20;", engine) 
rental_info = pd.DataFrame(rental_info)
```

```python
rental_info 
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
day_type = pd.read_sql_query("SELECT *, MONTH(rental_date) AS 'month', DAY(rental_date) AS 'weekday', DAYOFWEEK(rental_date) as 'day_type' FROM sakila.rental;", engine)
day_type = pd.DataFrame(day_type)
```
